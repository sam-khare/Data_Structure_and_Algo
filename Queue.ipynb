```python
from collections import deque
class Queue:
  def __init__(self):
    self.buffer = deque()

  def enqueue(self, val):
    self.buffer.appendleft(val)

  def dequeue(self):
    return self.buffer.pop()

  def is_empty(self):
    return len(self.buffer)==0

  def size(self):
    return len(self.buffer)
```

```python
pq = Queue()
pq.enqueue ({
    'company' : 'wal mart',
    'timestamp' : '15 apr, 11:01 AM',
    'price' : 131.10
})

pq.enqueue ({
    'company' : 'wal mart',
    'timestamp' : '15 apr, 11:02 AM',
    'price' : 132.12
})

pq.enqueue ({
    'company' : 'wal mart',
    'timestamp' : '15 apr, 11:03 AM',
    'price' : 135.10
})
```

#Multithreading

```python
import time
import threading

def calc_square(num):
  print("calculate square of numbers")
  for n in num:
    time.sleep(0.2)
    print('square',n*n)


def calc_cube(num):
  print("calculate cube of numbers")
  for n in num:
    time.sleep(0.2)
    print('cube',n*n*n)

arr = [2,3,8,9]

t = time.time()

'''calc_square(arr)
calc_cube(arr)'''

t1 = threading.Thread(target = calc_square,args=(arr,))
t2 = threading.Thread(target = calc_cube,args=(arr,))

t1.start()
t2.start()

t1.join()
t2.join()

print("done in..: ",time.time()-t)
```

```python
from collections import deque
class Queue:
  def __init__(self):
    self.buffer = deque()

  def enqueue(self, val):
    self.buffer.appendleft(val)

  def dequeue(self):
    return self.buffer.pop()

  def is_empty(self):
    return len(self.buffer)==0

  def size(self):
    return len(self.buffer)



```

```python
food_order_queue = Queue()

def placeOrder(orders):
  print('placing order..',orders)
  for d in orders:
    food_order_queue.enqueue(d)
    time.sleep(0.5)

def serveOrder():
  time.sleep(1)
  while True:
    order = food_order_queue.dequeue()
    print('Now serving',order)
    time.sleep(2)
```

```python
orders = ['pizza','samosa','pasta','biryani','burger']
t1 = threading.Thread(target = placeOrder,args=(orders,))
t2 = threading.Thread(target = serveOrder)
```

```python
t1.start()
t2.start()

t1.join()
t2.join()

```
