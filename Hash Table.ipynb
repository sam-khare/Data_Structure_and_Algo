```python
class HashTable:
  def __init__(self):
    self.MAX = 10
    self.arr = [[] for i in range(self.MAX)]

  def get_hash(self,key):
    h = 0
    for char in key:
      h+=ord(char)
    return h % self.MAX

  def __setitem__(self,key,val):
    h = self.get_hash(key)
    found = False
    for idx,element in enumerate(self.arr[h]):
      if len(element)==2 and element[0] == key:
        self.arr[h][idx] = (key,val)
        found =True
        break
    if not found:
      self.arr[h].append((key,val))

  def __getitem__(self,key):
    h = self.get_hash(key)
    for element in self.arr[h]:
      if element[0] == key:
        return element[1]


  def __delitem__(self,key):
    h = self.get_hash(key)
    for index,element in enumerate(self.arr[h]):
      if element[0] == key:
        del self.arr[h][index]
```

```python
t = HashTable()
```

```python
t['march 6'] = 155
t['march 8'] = 169
t['march 9'] = 65
t['march 17'] = 78
```

```python
word_count = {}
with open('poem.txt','r') as w:
  for line in w:
    tokens = line.split(' ')
    for token in tokens:
      token = token.replace('\n','')
      if token in word_count:
        word_count[token]+=1
      else:
        word_count[token] = 1

word_count
```
